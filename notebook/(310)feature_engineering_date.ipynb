```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset/train_date.csv')
```

```python
df
```

```python
features = df.drop(['Id'], axis=1).columns.values
```

```python
def orgainize(features):
    line_features = {}
    station_features = {}
    lines = set([f.split('_')[0] for f in features])
    stations = set([f.split('_')[1] for f in features])
    
    for l in lines:
        line_features[l] = [f for f in features if l+'_' in f]
        
    for s in stations:
        station_features[s] = [f for f in features if s+'_' in f]
        
            
    return line_features, station_features
```

```python
line_features, station_features = orgainize(features)
```

### Feature Engineering(Station)

```python
def make_features_station(df):
    new_features = pd.DataFrame({})
    station_duration_list =[]
    for s in station_features.keys():
        station_data = df[station_features[s]]
        col = s+'_first'
        new_features[col] = station_data.min(axis=1)
        col = s+'_last'
        new_features[col] = station_data.max(axis=1)
        col = s+'_duration'
        new_features[col] = station_data.max(axis=1) - station_data.min(axis=1)
        station_duration_list.append(col)
    return new_features, station_duration_list
```

```python
data_s,station_duration_list = make_features_station(df)
```

```python
data_s
```

```python
def make_features_station2(df):
    new_features = pd.DataFrame({})
    for s in station_duration_list:
        station_data = df[s]
        col = s+'_prev_diff'
        new_features[col] = station_data - station_data.shift(1)
        col = s+'_next_diff'
        new_features[col] = station_data - station_data.shift(-1)
    return new_features
```

```python
data_s_2= make_features_station2(data_s)
```

```python
data_s_2
```

### Feature Engineering(Line)

```python
def make_features_line(df):
    new_features = pd.DataFrame({})
    line_duration_list =[]
    for l in line_features.keys():
        line_data = df[line_features[l]]
        col = l+'_first'
        new_features[col] = line_data.max(axis=1)
        col = l+'_last'
        new_features[col] = line_data.min(axis=1)
        col = l+'_duration'
        new_features[col] = line_data.max(axis=1) - line_data.min(axis=1)
        line_duration_list.append(col)
    return new_features,line_duration_list
```

```python
data_l,line_duration_list= make_features_line(df)
```

```python
data_l
```

```python
def make_features_line2(df):
    new_features = pd.DataFrame({})
    for l in line_duration_list:
        line_data = df[l]
        col = l+'_prev_diff'
        new_features[col] = line_data - line_data.shift(1)
        col = l+'_next_diff'
        new_features[col] = line_data - line_data.shift(-1)
    return new_features
```

```python
data_l_2= make_features_line2(data_l)
```

```python
data_l_2
```

```python
df_FE_date =pd.concat([df['Id'],data_l,data_l_2,data_s,data_s_2],axis=1) 
```
